```python
def dist(p1, p2):
    return sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
    
df = pd.read_excel('0.xls')
array = df.to_numpy()
x = [el[0] for el in array]
y = [el[1] for el in array]

fig, axs = plt.subplots(2, 1, figsize = (5, 5))
axs[0].scatter(x, y)
axs[0].grid()

arr = [[], []]
for [x, y] in array:
    if 3 * y < 2 * x + 18:
        arr[0].append([x, y])
    else:
        arr[1].append([x, y])

colors = ['green', 'red', 'blue']
axs[1].grid()
ansx = 0
ansy = 0
for i in [0, 1]:
    best_p = [0, 0]
    best_dist = 1e9
    for p1 in arr[i]:
        cur_dist = 0
        for p2 in arr[i]:
            cur_dist += dist(p1, p2)
        if cur_dist < best_dist:
            best_dist = cur_dist
            best_p = p1
    ansx += best_p[0]
    ansy += best_p[1]
    
    x = [el[0] for el in arr[i]]
    y = [el[1] for el in arr[i]]
    axs[1].scatter(x, y, color = colors[i])
    
print(ansx / 2, ansy / 2)
```
